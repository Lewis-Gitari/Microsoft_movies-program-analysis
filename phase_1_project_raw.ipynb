```python
#lets name our dataframe with movies produced after 2000 as df2
df2 = above_2000_movies
```

```python
#Lets show microsoft the average budget for movie production upwards of the year 2000
print("The average budget is:",df2['Budget (in $)'].mean())
```

```python
#Lets show microsoft the perfomance between the different production houses.
Distributor_perfomances = df2.groupby('Distributor')['World Wide Sales (in $)'].mean()
Distributor_perfomances = Distributor_perfomances.apply(lambda x:'{:,.0f}'.format(x))
Distributor_perfomances
```

```python
plt.figure(figsize=(20,6))
plt.plot(Distributor_perfomances.index, Distributor_perfomances.values, marker='o',linestyle='-')
plt.title('Sales by Distributor')
plt.xlabel('Distributor')
plt.ylabel('World Wide Sales (in $)')
plt.grid(True)
plt.show()
```

Ideally i wanted a graph to show the different Distributors and their sales but it's abit murky lets try something different

```python
#create a dataframe representing the Distributor vs their cumulative worldwide sales.
df_distributor_perf = pd.DataFrame({'Distributor': Distributor_perfomances.index, 'Sales': Distributor_perfomances.values})
df_distributor_perf = df_distributor_perf.sort_values(by='Sales',ascending=False)
df_distributor_perf = df_distributor_perf.reset_index(drop=True)
df_distributor_perf
```

```python
#lets try figure out which Distributor had the highest Return On Investment.
#First lets create a table combining both the budget and world wide sales.
roi_df = df2.groupby('Distributor')[['Budget (in $)', 'World Wide Sales (in $)']].sum()
roi_df.sort_values(by='Budget (in $)',ascending=False)
roi_df.reset_index(inplace=True)
roi_df
```

```python
import numpy as np
```

```python
roi_df['Budget (in $)'] = roi_df['Budget (in $)'].abs().astype('int64')
#I realised i had some negative values soo i dropped them using the below line.
roi_df = roi_df[roi_df['Budget (in $)'] >=0]
roi_df
```

Some of the values have large numbers so that's why we have some values starting with a '-' sign.
So no need for alarm.

```python
#I intend to get the percentage Return on Investment percentage
#Now let's add a column depicting the Return On Investment percentage.
roi_df['%-RoI'] = ((roi_df['World Wide Sales (in $)'] - roi_df['Budget (in $)']) / roi_df['Budget (in $)'] * 100)
roi_df
```

```python
roi_df.info()
```

```python
#Lets plot a graph showing the Distributors RoI
plt.figure(figsize=(10,8))
plt.barh(roi_df['Distributor'],roi_df['%-RoI'], color='skyblue')
plt.xlabel(' %-RoI')
plt.title('Distributors RoI graph')
plt.gca().invert_yaxis()
plt.show()
```

```python
df3 = df2.copy()
```

```python
#lets create a Dataframe so as we can analyse the effect of release date on sales.
#Lets concentrate on movies released after year 2000.
df3.loc[:,'Release Date'] = pd.to_datetime(df3['Release Date'])
df3.loc[:,'Release Month'] = df3['Release Date'].dt.month

df3
```

```python
monthly_sales = df3.groupby('Release Month')['World Wide Sales (in $)'].sum()
monthly_sales = pd.DataFrame(monthly_sales)
monthly_sales.columns = ['World Wide Sales (in $)']
monthly_sales
```

```python
#Lets get visual image of the effect of month on sales, irrespective of the Distributor.
plt.figure(figsize=(10,6))
plt.barh(df3['Release Month'],df3['World Wide Sales (in $)'] , color='red')
plt.xlabel('World Wide Sales (in $)')
plt.title('Release month effect on World Wide Sales (in $)')
plt.gca().invert_yaxis()
plt.show()
```
